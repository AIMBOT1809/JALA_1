```python
num = 7
flag = True

if num <= 1:
    flag = False
else:
    for i in range(2, num):
        if num % i == 0:
            flag = False
            break

if flag:
    print("Prime Number")
else:
    print("Not a Prime Number")
```

```python
num = 121
temp = num
reverse = 0

while temp > 0:
    digit = temp % 10
    reverse = reverse * 10 + digit
    temp //= 10

if num == reverse:
    print("Palindrome")
else:
    print("Not a Palindrome")
```

4. ARRAYS

```python
def sum_array(arr):
    return sum(arr)
print(sum_array([1,2,3,4,5]))
```

```python
def average_array(arr):
    return sum(arr) / len(arr)
print(average_array([10,20,30,40]))
```

```python
arr=[10,20,30,40]
element = 30
if element in arr:
    print("Index:", arr.index(element))
else:
    print("Element not found")
```

```python
def contains_value(arr,value):
    return value in arr
print(contains_value([1,2,3,4],3))
```

```python
def remove_element(arr,value):
    if value in arr:
        arr.remove(value)
    return arr
print(remove_element([1,2,3,4],3))
```

```python
def copy_array(arr):
    return arr.copy()
print(copy_array([5,6,7]))
```

```python
def insert_element(arr,pos,value):
    arr.insert(pos, value)
    return arr
print(insert_element([1,2,4],2,3))
```

```python
def min_max(arr):
    return min(arr), max(arr)
print(min_max([10,5,20,8]))
```

```python
def reverse_array(arr):
    return arr[::-1]
print(reverse_array([1, 2, 3, 4]))
```

```python
def find_duplicates(arr):
    duplicates=[]
    for i in arr:
        if arr.count(i) > 1 and i not in duplicates:
            duplicates.append(i)
    return duplicates
print(find_duplicates([1, 2, 3, 2, 4, 1]))
```

```python
arr1 = [1,2,3,4]
arr2 = [3,4,5,6]
common = list(set(arr1) & set(arr2))
print(common)
```

```python
def remove_duplicates(arr):
    return list(set(arr))
print(remove_duplicates([1,2,2,3,4,4]))
```

```python
def second_largest(arr):
    arr=list(set(arr))
    arr.sort()
    return arr[-2]
print(second_largest([10,20,30,40,50]))
```

```python
def second_largest_alt(arr):
    first = second = float('-inf')
    for num in arr:
        if num > first:
            second=first
            first=num
        elif num > second and num!=first:
            second=num
    return second
print(second_largest_alt([10, 20, 30, 40, 50]))
```

```python
def count_even_odd(arr):
    even=odd=0
    for num in arr:
        if num%2==0:
            even += 1
        else:
            odd += 1
    return even, odd
print(count_even_odd([1,2,3,4,5,6]))
```

```python
def difference(arr):
    return max(arr) - min(arr)
print(difference([10, 5, 20, 8]))
```

```python
def contains_both(arr):
    return 12 in arr and 23 in arr
print(contains_both([12, 5, 23, 9]))
```

```python
def remove_duplicates_new(arr):
    new_arr=[]
    for num in arr:
        if num not in new_arr:
            new_arr.append(num)
    return new_arr
print(remove_duplicates_new([1,2,2,3,4,4]))
```
