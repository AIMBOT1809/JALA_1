```python
#30. Perform Singular Value Decomposition (SVD) on a 3x3 matrix and reconstruct the original matrix using the SVD components.

A = np.array([[4, 0, 3],
              [2, 1, 5],
              [0, 6, 7]])

# Perform SVD
U, S, Vt = np.linalg.svd(A)

# Convert singular values into a diagonal matrix
S_matrix = np.zeros((3, 3))
np.fill_diagonal(S_matrix, S)

# Reconstruct the original matrix
A_reconstructed = U @ S_matrix @ Vt

# Display results
print("Original Matrix:\n", A)
print("\nU Matrix:\n", U)
print("\nSingular Values:\n", S)
print("\nV Transpose Matrix:\n", Vt)
print("\nReconstructed Matrix:\n", A_reconstructed)
```

```python
q1=df["salary"].quantile(0.25)
q3=df["salary"].quantile(0.75)
IQR=q3-q1
print(IQR)

ll=q1-1.5*IQR
ul=q3+1.5*IQR
outlier_mask=(df["salary"]<ll)|(df["salary"]>ul)
median_value = df["salary"].median()
df.loc[outlier_mask, "salary"] = median_value
df
```
